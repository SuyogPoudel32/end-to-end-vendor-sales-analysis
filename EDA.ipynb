```python
query = """WITH PurchasePriceSummary AS (
    SELECT
        p.VendorNumber,
        p.VendorName,
        p.Brand,
        p.PurchasePrice,
        pp.Price AS ActualPrice,
        pp.Volume,
        p.Description,
        SUM(p.Quantity) AS TotalPurchaseQuantity,
        ROUND(SUM(p.Dollars),2) AS TotalPurchaseDollar
    FROM purchases p
    JOIN purchase_prices pp USING (Brand)
    GROUP BY
        p.VendorNumber,
        p.VendorName,
        p.Brand,
        pp.Price,
        p.PurchasePrice,
        pp.Volume,
        p.Description
),

VendorInvoiceSummary AS (
    SELECT
        vi.VendorNumber,
        SUM(vi.Freight) AS TotalFreight
    FROM vendor_invoice vi
    GROUP BY vi.VendorNumber
),

SalesSummary AS (
    SELECT
        s.VendorNo,
        s.Brand,
        s.VendorName,
        SUM(s.SalesQuantity) AS TotalSalesQuantity,
        ROUND(SUM(s.SalesDollars),2) AS TotalSalesDollars,
        ROUND(SUM(s.SalesPrice),2) AS TotalSalesPrice,
        ROUND(SUM(s.ExciseTax)) AS ExciseTax
    FROM sales s
    GROUP BY
        s.VendorNo,
        s.VendorName,
        s.Brand
)

SELECT
    pps.VendorNumber,
    pps.VendorName,
    pps.Brand,
    pps.Volume,
    pps.Description,
    pps.ActualPrice,
    pps.PurchasePrice,
    pps.TotalPurchaseQuantity,
    pps.TotalPurchaseDollar,

    ss.TotalSalesQuantity,
    ss.TotalSalesDollars,
    ss.TotalSalesPrice,
    ss.ExciseTax,

    vis.TotalFreight

FROM PurchasePriceSummary pps

LEFT JOIN SalesSummary ss
    ON pps.VendorNumber = ss.VendorNo
   AND pps.Brand = ss.Brand

LEFT JOIN VendorInvoiceSummary vis
    ON pps.VendorNumber = vis.VendorNumber;
"""

vendor_sales_summay = pd.read_sql_query(query,connection)
```

```python
vendor_sales_summay_copy = vendor_sales_summay.copy()
```

```python
vendor_sales_summay_copy.fillna(0,inplace=True)
```

```python
def change_dtype(col_name,d_type):
    vendor_sales_summay_copy[col_name] = (vendor_sales_summay_copy[col_name]).astype(d_type)
```

```python
for col,d_type in zip(["Volume","Volume","TotalPurchaseQuantity","TotalSalesQuantity"],["float","int","int","int"]):
    change_dtype(col,d_type)
```

```python
vendor_sales_summay_copy.info()
```

```python
def strip_data(col_name):
    vendor_sales_summay_copy[col_name] = vendor_sales_summay_copy[col_name].str.strip()
```

```python
for col in ["VendorName","Description"]:
    strip_data(col)
```

```python
vendor_sales_summay_copy["GrossProfit"] = vendor_sales_summay_copy["TotalSalesDollars"] - vendor_sales_summay_copy["TotalPurchaseDollar"]
```

```python
vendor_sales_summay_copy["ProfitMargin"] = (vendor_sales_summay_copy["GrossProfit"] / vendor_sales_summay_copy["TotalSalesDollars"])*100
```

```python
vendor_sales_summay_copy["StockTurnover"] = vendor_sales_summay_copy["TotalSalesQuantity"] / vendor_sales_summay_copy["TotalPurchaseQuantity"]
```

```python
vendor_sales_summay_copy["SalesToPurchaseRatio"] = vendor_sales_summay_copy["TotalSalesDollars"] / vendor_sales_summay_copy["TotalPurchaseDollar"]
```

```python
vendor_sales_summay_copy.columns
```

```python
vendor_sales_summay_copy = vendor_sales_summay_copy.replace([np.inf,-np.inf],np.nan)
```

```python
vendor_sales_summay_copy = vendor_sales_summay_copy.fillna(0)
```
